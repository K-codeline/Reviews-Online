```python
import json
from string import punctuation
from pprint import pprint

jsonlistn=[]
#An empty list is created for us to add the negative reviews in the list.
for i in range (1,1537+1):
#This loop is from 1 to the last number of discussions+1. 
#As we are not starting from zero, the loop will run one less time if we do not plus one.
    jsonpathn = ('C:/Users/Aaron/Desktop/reviews/reviews/Negative/discussions_'+str(format(i,"07d"))+'.json')
    jsondatan= open(jsonpathn,encoding="utf8")
    #"utf-8" is needed as we have to change the encoding for our file to run.
    jsondatan=json.load(jsondatan)
    jsondatan=jsondatan['text']
    #jtp['text'] refers to only the 'text' part in each json file.
    #pprint(jsondatan)      # to show the output for the text file.

    
#Part 2
    jsondatan=jsondatan.lower()
    #jsondatan will be in lower case.
    for p in punctuation:
        jsondatan=(jsondatan).replace(p,"")
        #The puncutation of jsondatapnis removed.
    jsondatan=(jsondatan).replace("·","")  
    #The bullet point ("·") is not considered a puncation. Therefore, the puncutation code is unable to remove it.
    #The bullet point is not a word, therefore it is necessary to be removed.
    jsondatan=jsondatan.split()
    #jsontan will become a list of the words of 'text' of the reviews.
    jsonlistn.extend(jsondatan)
    #Since jsondatan is a list, we need to use extend instead of append so that we can remove the square brackets.
    #We made it into a list so that it will be more convenient if we want to have the data in another cell.
    #Without the list, when we are doing the next part, a loop is needed.
```

```python
commonwords = set(jsonlistp).intersection(set(jsonlistn))
#This is used to find the common words between the 2 lists.
#Set helps us to make the elements in jsonlist to be unqiue.
#With set, we can use intersection to find the common words between the 2 lists.
```

```python
#Remove common words from Positive.
UniquePos=[]
#UniquePos represents the unique positive words that are not inisde the negative reviews.
i=0
for jsonlistp[i] in jsonlistp:
    if jsonlistp[i] not in commonwords:
    #jsonlistp[i] represents the elements inside the list jsonlistp.
        UniquePos.extend(list(jsonlistp[i].split()))
        #If the word is not in commonwords, then we extend in into a new list, UniquePos.
        #Without ".split", they give you the most common letter.
        i+=1
```

```python
#ptop is the top 150 unique positive words.
from collections import Counter
ptop=Counter(UniquePos).most_common(150)
#".most_common(150)" gives me the most common 150 element in the list UniquePos.
ptop
```

```python
#Remove common words from Negative.
UniqueNeg=[]
#UniqueNeg represents the unique negative words that are not inisde the negative reviews.

for jsonlistn[i] in jsonlistn:
    if jsonlistn[i] not in commonwords:
        #jsonlistp[i] represents the elements inside the list jsonlistp.
        UniqueNeg.extend(list(jsonlistn[i].split()))
        #If the word is not in commonwords, then we extend in into a new list, UniqueNeg.
        #Without ".split", the output will be the most common individual letter.
```

```python
#ntop is the top 150 unique negative words.
from collections import Counter
ntop=Counter(UniqueNeg).most_common(150)
#".most_common(150)" gives me the most common 150 element in the list UniqueNeg.
ntop
```

```python
ptopnew=[]
ntopnew=[]
# 2 empty lists are created so that we can add the top 150 words without the numbers inside.
def ne(word,li):
#A function is defined so that we are able to have the numbers which serves as a counter in a new list.
    for i in range(0,150):
        hi=str(word[i]).split(",")[0].split("'")[1].split("'")[0].strip()
        #str is needed as we cannot split a list.
        #hi refers to the words that do not include the numbers.
        #The split process includes splitting before the comma(","), after the first apostrophe and before the last apostrophe.
        li.append(hi)
        #The words are appended into a list.
```

```python
ne(ptop,ptopnew)
#ne is applied. The top 150 positive words which does not includes the numbers is inside the ptopnew, a list.
ne(ntop,ntopnew)
#ne is applied. The top 150 negative words which does not includes the numbers is inside the ntopnew, a list.
```

```python
#Part 3
import json
from string import punctuation
from pprint import pprint

reviewpos=[]
#An empty list is created for us to add the overall scores of each review into the list.
for i in range (1,3290+1):
#This loop is from 1 to the last number of discussions+1. 
#As we are not starting from zero, the loop will run one less time if we do not plus one.
    jpath = ('C:/Users/Aaron/Desktop/reviews/reviews/Test_Pos/discussions_'+str(format(i,"07d"))+'.json')
    jtp= open(jpath,encoding="utf8")
    #"utf-8" is needed as we have to change the encoding for our file to run.
    jtp=json.load(jtp)
    jtp=jtp['text']
    #jtp['text'] refers to only the 'text' part in each json file.
    jtp=jtp.lower()
    #jtp will be in in lower case.
    for p in punctuation:
        jtp=(jtp).replace(p,"")
    #jtp will no longer have any puncutation.
    jtp=jtp.split()
    #jtp will become a list of the words of 'text' of the reviews.

    positive_review_p=0
    #Number of positive Reviews inside Test_Pos file.
    negative_review_p=0
    #Number of negative Reviews inside Test_Pos file.
    review_p=0
    #Number of neutral Reviews inside Test_Pos file.
    for i in jtp:
    #for every word in the list jtp.
        if i in ptopnew:
            positive_review_p+=1
            #positive_review_p will increase by one if the word is inside the top 150 positive words.
        elif i in ntopnew:
            negative_review_p+=1
            #negative_review_p will increase by one if the word is inside the top 150 negative words.
        else:
            review_p+=1
            #review_p will increase by one if the word is not inside any of the top 150 list.
    overall_review_p= positive_review_p - negative_review_p
    #Overall score of each review. 
    reviewpos.append(overall_review_p)
    #The scores for each review is appended inside the list reviewpos.
```

```python
Review_PP=0
#Number of positive review in Test_Pos folder.
Review_PN=0
#Number of negative review in Test_Pos folder.
Review_PNeutral=0
#Number of neutral review in Test_Pos folder.
for i in reviewpos:
    if i >0:
        Review_PP+=1
    #Review_PP will increase by one if the score is more than zero (i.e. 1/2/3/4).
    elif i<0:
        Review_PN+=1
    #Review_PN will increase by one if the score is less than zero (i.e. -1/-2/-3/-4).
    elif i==0:
        Review_PNeutral+=1
    #Review_PP will increase by one if the score is zero.
    
    
print("Test_Pos:")
print("Number of reviews predicted as positive:",Review_PP)
print("Number of reviews predicted as negative:",Review_PN)
print("Others:",Review_PNeutral)
#The results are printed.
```
